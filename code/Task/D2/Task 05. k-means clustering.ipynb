```python
# *Solutions Examples
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import metrics
# Generating data sets
X, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [0,0], [1,1], [2,2]], cluster_std=[0.4, 0.2, 0.2, 0.2], 
                  random_state =9)
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()

#Loop through different cluster numbers to generate clusters and evaluate the clustering results
# plt.subplots_adjust(left=.02, right=.98, bottom=.096, top=.96, wspace=.1,
#                     hspace=.1)
plt.figure(figsize=(10, 6))
for index, k in enumerate((2,3,4,5)):
    plt.subplot(2,2,index+1)
    y_pred = KMeans(n_clusters=k, random_state=9).fit_predict(X)
    s_Score=metrics.silhouette_score(X, y_pred, metric='euclidean', sample_size=None, random_state=None)  

    plt.scatter(X[:, 0], X[:, 1], c=y_pred)
    plt.text(.9, .1, ('k=%d, s_score: %.2f' % (k,s_Score)),
                 transform=plt.gca().transAxes, size=10,
                 horizontalalignment='right')
plt.show()
```
